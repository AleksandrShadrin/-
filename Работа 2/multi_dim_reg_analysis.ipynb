#### Сформируем таблицу на основе статистических данных

```python
def convert_to_int(string):
    try:
        return int(string.strip())
    except Exception:
        return np.NaN
```

```python
import urllib3
from bs4 import BeautifulSoup

http = urllib3.PoolManager()

url = 'https://www.gks.ru/bgd/regl/B04_14/IssWWW.exe/Stg/d010/i010030r.htm'
req = http.request('GET', url)
soup = BeautifulSoup(req.data.decode(encoding='cp1251'))
table = soup.findAll('tr')
table = [[j.get_text(" ",strip=True) for j in i.children if j.get_text(" ",strip=True) != ''] for i in table]
table

df = pd.DataFrame(columns=['Регион'] + table[0][0:3]+['растениеводства', 'животноводства']+table[0][3:-1], data=table[2:-2])
df = df.set_index('Регион')

for column in df.columns:
    df[column] = df[column].apply(convert_to_int)

df = df.dropna()
df.head(10)
```
